```python
#import libraries
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import  datetime as dt
```

```python
#Load the dataset into data frame
df = pd.read_csv ('supermarket_sales - Sheet1.csv')
df
```

```python
#checking number or rows and columns
df.shape
```

```python
#how does the data look?
df.sample(5)
```

```python
#basic information
df.info
```

```python
#checking the missing values
df.isnull().sum()
```

```python
#checking the column types
df.dtypes
```

```python
#from this we know, Date and Time is in object data type, for better analysis we will change it to DateTime
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])
df.dtypes
```

```python
#Creating new columns bsed on Date columns
df["Days"]=df["Date"].dt.day_name()
df["Month"]=df["Date"].dt.month_name()
df["weekday"]=df.Date.dt.weekday
```

```python
#Creating new columns bsed on Time columns
df["Time"].mode()
df["Hour"]=df["Time"].dt.hour
```

```python
#Branch A is in city Yangon
#Branch B is in city Mandalay
#Branch c is in city Naypyitaw
#so we will keep one of them and drop other and invoiceid is irrelevant to our data so we will drop that too
df.drop(columns=["Invoice ID","Branch"],inplace=True)
```

```python
#Descriptive statistical summary of object data
df.describe(include="object").T
```

```python
#find out the total sales and gross income of the Supercenter
df["Total"].sum()
```

```python
df["gross income"].sum()
```

```python
#find out the total sales of each City
df.groupby("City")["Total"].agg(["sum","mean","min","max"])
```

```python
#Shows how does the data look now?
df.sample(5)
```

```python
#Plot pairwise relationships in a dataset.
sns.pairplot(df)
plt.show()
```

```python
#Checking Correlation with heatmap
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
#checking all the product lines and its count
fig = px.histogram(df, x='Product line', color="City", text_auto=True)
fig.show()
```

```python
#Checking which gender have the total bill price
fig = px.pie(df, values='Total', names='Gender',hole=0.5)
fig.show()
```

```python
#Checking if total price affect on the rating of the supercenter in different cities
fig = px.scatter(df, x="Total", y="Rating",animation_frame="City", size="Total", color="City")
fig.show()
```

```python
#Checking if all the product lines have the same range of price
fig = px.box(df,  x="Product line", y="Unit price", color="City")
fig.show()
```

```python
#Checking if Customer type affect on the rating of the supercenter in different cities
plt.figure(figsize=(15,5))
sns.countplot(x="Rating",data=df,hue="Customer type");
plt.xticks(rotation=90);
plt.show()
```

```python
#Checking if there is a difference between women and men in product line interests
plt.figure(figsize=(12,5))
sns.barplot(data=df,x="Gender",y="Total",hue="Product line",estimator=sum)
plt.show()
```

```python
#Checking the total Sales by Hour of Day
sns.relplot(kind = 'line', data = df, x ='Hour', y = 'Total', aspect = 2, hue = 'Days', ci = False,estimator=sum)
plt.xlabel('Hour of Day')
plt.title('Sales by Hour of Day (24 hrs)')
plt.show()
```

```python
#Checking if there is a diffrence between member and normal customer in total price
sns.stripplot(y='Total',x='Customer type',data=df)
plt.show()
```

```python
#Checking how many votes in every rating
plt.figure(figsize = (12,6))
df['Rating'].value_counts().plot(kind="bar")
plt.show()
```

```python
#Checking the relationship between total price and rating?
sns.jointplot(
    data=df,
    x="Total", y="Rating", hue="City",
    kind="kde")
plt.show()
```

```python
#Checking the total profits in the 3 months among all cities
monthly_sales = pd.pivot_table(df, values = "gross income", columns = "City", index = "Month", aggfunc= 'sum')
monthly_sales.plot()
plt.show()
```

```python
#Checking the most used payment and its total price
fig = px.pie(df, values='Total', names='Payment')
fig.show()
```
